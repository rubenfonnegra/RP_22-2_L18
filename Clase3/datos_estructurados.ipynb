# Estructurando im√°genes en Python

```python
import os
import cv2
from PIL import Image
import matplotlib.pyplot as plt
```

```python
path = "/content/drive/MyDrive/Docencia/Patrones/2022-2/L18/patrones/Clase3/sound_image/"

list_ext = [".jpg", ".png", ".jpeg", ".tif", ".tiff", ".bmp", ".gif"]
list_files = []

for file_ in os.listdir(path): 
  # print(os.path.splitext(file_))
  if os.path.splitext(file_)[1].lower() in list_ext: 
    #print(file_)
    list_files.append(path + file_)

print (list_files)
```

```python
list_images = []

for image in list_files: 
  #print(image)
  image = cv2.imread(image)

  ##--- Preprocesamiento adicional ---##
  image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
  image = cv2.resize(image, (256,256))

  list_images.append(image)

print(len(list_images))
```

```python
for image in list_images: 
  plt.imshow(image)
  plt.show()
```
